```python
from statistics import mean, mode, median, median_low, median_high, variance, stdev
import random
intList = random.sample(range(0,100), 20)
print(intList)
floatList = [2-5,6.5,3.4,6.7,7.0,8.3]
print(floatList)
negList = random.sample(range(-100,-1), 20)
print(negList)
numList = random.sample(range(-100,100), 20)
print(numList)
print("MEAN")
print("the mean of the intList is %.2f"%mean(intList))
print("the mean of the intList is %.2f"%mean(floatList))
print("the mean of the intList is %.2f"%mean(negList))
print("the mean of the intList is %.2f"%mean(numList))
print("MODE")
print("the mode of the intList is %.2f"%mode(intList))
print("the mode of the intList is %.2f"%mode(floatList))
print("the mode of the intList is %.2f"%mode(negList))
print("the mode of the intList is %.2f"%mode(numList))
print("MEDIAN")
print("the median of the intList is %.2f"%median(intList))
print("the median of the intList is %.2f"%median(floatList))
print("the median of the intList is %.2f"%median(negList))
print("the median of the intList is %.2f"%median(numList))
print("MEDIAN_LOW")
print("the median_low of the intList is %.2f"%median_low(intList))
print("the median_low of the intList is %.2f"%median_low(floatList))
print("the median_low of the intList is %.2f"%median_low(negList))
print("the median_low of the intList is %.2f"%median_low(numList))
print("MEDIAN_HIGH")
print("the median_high of the intList is %.2f"%median_high(intList))
print("the median_high of the intList is %.2f"%median_high(floatList))
print("the median_high of the intList is %.2f"%median_high(negList))
print("the median_high of the intList is %.2f"%median_high(numList))
print("VARIANCE")
print("the variance of the intList is %.2f"%variance(intList))
print("the variance of the intList is %.2f"%variance(floatList))
print("the variance of the intList is %.2f"%variance(negList))
print("the variance of the intList is %.2f"%variance(numList))
print("STDEV")
print("the stdev of the intList is %.2f"%stdev(intList))
print("the stdev of the intList is %.2f"%stdev(floatList))
print("the stdev of the intList is %.2f"%stdev(negList))
print("the stdev of the intList is %.2f"%stdev(numList))
```
